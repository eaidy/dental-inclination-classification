```python
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
""" PRE-PROCESSING AREA """


image = cv2.imread('../image.jpg')
image = cv2.resize(image, (400, 300))
image_array = np.array(image)
image_array = image_array / 255.0

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.colorbar()
plt.show()
```

```python
img_height, img_width, img_pixel = image_array.shape
reduced_image_array = np.zeros((img_height, img_width, img_pixel))

def reduce_pixel_matrix(pixel_matrix):
    return np.array([
        pixel_matrix[0],
        pixel_matrix[1],
        pixel_matrix[2]
    ])

for i in range(0, img_height):
    for j in range(0, img_width):
        reduced_value = reduce_pixel_matrix(image_array[i][j])
        reduced_image_array[i][j] = reduced_value

plt.imshow(reduced_image_array, cmap='gray')
plt.colorbar()
plt.show()
```
